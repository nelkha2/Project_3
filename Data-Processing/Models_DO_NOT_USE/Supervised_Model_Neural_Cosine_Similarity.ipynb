# Embedding - Word2Vec

```python
# importing all necessary modules 
from nltk.tokenize import sent_tokenize, word_tokenize 
import nltk
import warnings 
import gensim 
from gensim.models import Word2Vec 
import multiprocessing 
```

```python
# Override warning notifications 
warnings.filterwarnings(action = 'ignore')

# Parse phrases into words
words_clean = final_data["words"]
words_clean = words_clean.replace("/n"," ")
# convert to acceptable format for tokenization
words_clean = str(words_clean)
words_clean
```

```python
# Tokenize

data = [] 
for i in sent_tokenize(words_clean): 
    temp = [] 
    
    #tokenize phrases into words 
    for j in word_tokenize(i): 
        temp.append(j.lower()) 
  
    data.append(temp)
data
```

```python
# Skip-Gram calculation 
model_gram = gensim.models.Word2Vec(data, min_count = 1, size = 100, window = 5, sg = 1)

# Print similarity score 
model_gram.similarity('shoe','sock')
```

```python
#Cosine calculation 
model_cosine = gensim.models.Word2Vec(data, min_count = 1,  size = 100, window = 5)

#Print similarity score 
model_cosine.similarity("address","labels")
```

```python
# Create & fit Word2Vec Model (Simple 2 layer Neural Network)
# window: number of words to consider to target from, n words from right and n words from left
# iter: epochs
# min_count: how many times a word needs to appear to be a component of the library
model_w2v = Word2Vec(data, size=10, window=100, min_count=1, iter=30, workers=multiprocessing.cpu_count())
```

```python
# Predictions 
input = "sneakers"
word_vectors = model_w2v.wv
results = word_vectors.similar_by_word(input)
results
```
