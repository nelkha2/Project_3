# Data Processing 

```python
import pandas as pd
import numpy as np
import re
```

```python
csv_file = "../Data/phrases.csv"
df = pd.read_csv(csv_file,encoding="utf-8")
```

```python
pd.concat([df[col].astype(str).str.lower() for col in df.columns], axis=1)
df.head()
```

```python
df['wordCount'] = [len(x.split()) for x in df['words'].tolist()]
```

```python
df.wordCount.value_counts()
```

```python
df[['word1','word2','word3','word4','word5','word6','word7']] = df.words.str.split(" ",expand=True,)
```

```python
df['word1_count']=df['word1'].str.len()
df['word2_count']=df['word2'].str.len()
df['word3_count']=df['word3'].str.len()
df['word4_count']=df['word4'].str.len()
df['word5_count']=df['word5'].str.len()
df['word6_count']=df['word6'].str.len()
df['word7_count']=df['word7'].str.len()
```

```python
from collections import Counter

df1 = pd.DataFrame([Counter(list(x)) for x in df['words']], index=df.index)
order = list('abcdefghijklmnopqrstuvwxyz')

d = {'a': "count_a",
     'b': "count_b",
     'c':"count_c",
     'd':"count_d",
     'e':"count_e",
     'f':"count_f",
     'g':"count_g",
     'h':"count_h",
     'i':"count_i",
     'j':"count_j",
     'k':"count_k",
     'l':"count_l",
     'm':"count_m",
     'n':"count_n",
     'o':"count_o",
     'p':"count_p",
     'q':"count_q",
     'r':"count_r",
     's':"count_s",
     't':"count_t",
     'u':"count_u",
     'v':"count_v",
     'w':"count_w",
     'x':"count_x",
     'y':"count_y",
     'z':"count_z"
     }

df2 = df1.fillna(0).astype(int).reindex(columns=order, fill_value=0).rename(columns=d)
df = df.join(df2)
```

```python
df.head()
```

```python
phrase_data = df
```

```python
phrase_data.columns
```

```python
phrase_data=phrase_data.drop(columns=['word1','word2','word3','word4','word5','word6','word7'])
```

```python
final_data= phrase_data.fillna(0)
```

# Model - SVM

```python
# Dependencies 
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import os
```

```python
# Assign X & y 
X = final_data.drop("words", axis=1)
y = final_data["words"]
```

```python
print(X.shape, y.shape)
```

```python
#Data Split: Train vs Test  
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Scale Data 
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Explore independent variables (X) 
X_train_scaled
```

```python
#Create Model 
from sklearn.svm import SVC 
model = SVC(kernel='linear')

#Fit Model (train data)
model.fit(X_train_scaled, y_train)
```

```python
# Predictions 
predictions = model.predict(X_test_scaled)
predictions
```

```python
# Evaluate / Validate Model - test data 

from sklearn.metrics import classification_report

print('Test Acc: %.3f' % model.score(X_test_scaled, y_test))
print(classification_report(y_test, predictions))
```
